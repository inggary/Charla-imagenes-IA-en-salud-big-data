```python
import kagglehub
from pathlib import Path
import shutil
import os

# Determinar la carpeta 'data' relativa al directorio de ejecución
cwd = Path.cwd()
if cwd.name == 'notebooks':
    data_dir = cwd.parent / 'data'
else:
    data_dir = cwd / 'data'
data_dir.mkdir(parents=True, exist_ok=True)

# Descargar la última versión (kagglehub devuelve la ruta al archivo descargado)
path = kagglehub.dataset_download("sovitrath/diabetic-retinopathy-224x224-gaussian-filtered")
print("Path devuelto por kagglehub:", path)

# Intentar mover el archivo devuelto a la carpeta data/ (si existe)
try:
    src = Path(path)
    if src.exists():
        dest = data_dir / src.name
        # Si el archivo ya existe en destino, lo sobrescribimos moviendo (shutil.move reemplaza)
        shutil.move(str(src), str(dest))
        print("Archivo movido a:", dest)
    else:
        # En algunos casos kagglehub ya descomprime/coloca archivos en una carpeta; imprimimos info para depuración
        print("La ruta devuelta no existe en el FS. Comprueba si kagglehub creó una carpeta en su lugar.")
except Exception as e:
    print("Error moviendo archivo a data/:", e)

print("Carpeta data usada:", data_dir)
```
